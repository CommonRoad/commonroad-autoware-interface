```python
# import functions to read xml file and visualize commonroad objects
import matplotlib.pyplot as plt
from commonroad.common.file_reader import CommonRoadFileReader
from commonroad.visualization.mp_renderer import MPRenderer

# generate path of the file to be opened
file_path = "../data/sample-map-cr/PRI_Barceloneta-2_2_T-1.xml"
# file_path = "/home/drivingsim/autoware/src/universe/autoware.universe/planning/tum_commonroad_planning/dfg-car/src/cr2autoware/data/ZAM_Over-1_1.xml"
# file_path = "sample-map-planning.xml"
# file_path = "/home/drivingsim/autoware/src/universe/autoware.universe/planning/tum_commonroad_planning/commonroad-scenario-designer/tests/map_conversion/test_maps/lanelet2/merging_lanelets_utm.xml"



# read in the scenario and planning problem set
scenario, planning_problem_set = CommonRoadFileReader(file_path).open()

# plot the planning problem and the scenario for the fifth time step
plt.figure(figsize=(30, 30))
rnd = MPRenderer()
scenario.draw(rnd, draw_params={'time_begin': 0})
planning_problem_set.draw(rnd)
rnd.render()
plt.show()
```
